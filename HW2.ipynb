```python
# -*- coding: utf-8 -*-
"""
Created on Tue Sep 24 17:33:11 2019

@author: erick
"""
import re
from datetime import datetime
import pandas as pd
import numpy as np
from random import randint
from faker import Faker
fake = Faker()
#creates dictionary from fake data using faker module
def input_data(x):
    person_data ={}
    birthday = {}
    for i in range(0, x):
        person_data[i]={}
        person_data[i]['email']= fake.email()
        person_data[i]['phone number']= fake.phone_number()
        person_data[i]['home address']= str(fake.address()).replace('\n',' ')
        person_data[i]['name']= str(fake.name())
        birthday[i] = datetime.strftime(fake.date_of_birth(minimum_age=18,maximum_age=80),format="%Y")
        person_data[i]['year born']= (birthday[0])
        person_data[i]['number of kids'] = np.random.randint(0,4)
        house = ['rent', 'own']
        person_data[i]['rent or own?'] = str(np.random.choice(house))
        incomes = list(range(10,200,10))
        person_data[i]['annual income'] = int(np.random.choice(incomes, 1))
        tickets = list(range(0,14))
        person_data[i]['speeding tickets in past year'] = int(np.random.choice(tickets, 1))
    return(person_data)

def main():
 
    # Enter number of persons
    c=input('How many number of persons would you like data from? integer only: ')
    try:
        nested_dict = input_data(int(c))
    except: raise UnboundLocalError("The Input is not an integer")
    df = pd.DataFrame.from_dict({i: nested_dict[i] for i in nested_dict.keys()},orient='index')
    print(df)
    print("The available columns are: \n" ,df.columns.values)
    a = input("Please enter desired columns from dataset: ")# Enter desired columns as strings
    try:
        word = len(re.findall(r'\w+', str.casefold(a)))
        if word == 1:
            ndf = pd.Series.to_frame(df[a])
            print(ndf)
        elif word > 1:
            la = a.split(",")
            ndf = df[list(la)]
            print(ndf)
    except: "The Input column is invalid"
    b = input("Please choose output file format (only json & csv): ")
    if b == 'json' or b == 'javascript object notation': ndf.to_json(path_or_buf = "C:\\Users\\erick\\Desktop\\out.json")
    elif b == 'csv' or b == 'comma separated values': ndf.to_csv(path_or_buf = "C:\\Users\\erick\\Desktop\\out1.csv")
    else: raise Exception(" The format you requested is invalid")
main()
```
